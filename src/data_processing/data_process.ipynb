```python
import pandas as pd
import numpy as np
df= pd.read_csv("./ObesityDataSet_raw_and_data_sinthetic.csv") # path
df.head()
```

```python
# binary categories to 0/1
from sklearn.preprocessing import LabelEncoder 
from numpy import array 

def bi_category(category):
    cate=array(category)
    encoder = LabelEncoder()
    cate_encoded =  encoder.fit_transform(cate) 
    category = cate_encoded
    return cate_encoded

df.Gender = bi_category(df.Gender)
df.family_history_with_overweight = bi_category(df.family_history_with_overweight)
df.FAVC = bi_category(df.FAVC)
df.SMOKE = bi_category(df.SMOKE)
df.SCC = bi_category(df.SCC)
df.head()
```

```python
# frequency categories
def freq_category(s):
    df.loc[df[s] == "no", s] = 0
    df.loc[df[s] == "Sometimes", s] = 1
    df.loc[df[s] == "Frequently", s] = 2
    df.loc[df[s] == "Always", s] = 3
freq_category('CAEC')
freq_category('CALC')
df.head()
```

```python
# MTRANS and label 
df.loc[df["MTRANS"] == "Automobile", "MTRANS"] = 0
df.loc[df["MTRANS"] == "Motorbike", "MTRANS"] = 1
df.loc[df["MTRANS"] == "Public_Transportation", "MTRANS"] = 2
df.loc[df["MTRANS"] == "Bike", "MTRANS"] = 3
df.loc[df["MTRANS"] == "Walking", "MTRANS"] = 4

df.loc[df["NObeyesdad"] == "Insufficient_Weight", "NObeyesdad"] = 0
df.loc[df["NObeyesdad"] == "Normal_Weight", "NObeyesdad"] = 1
df.loc[df["NObeyesdad"] == "Overweight_Level_I", "NObeyesdad"] = 2
df.loc[df["NObeyesdad"] == "Overweight_Level_II", "NObeyesdad"] = 3
df.loc[df["NObeyesdad"] == "Overweight_Level_III", "NObeyesdad"] = 4
df.loc[df["NObeyesdad"] == "Obesity_Type_I", "NObeyesdad"] = 5
df.loc[df["NObeyesdad"] == "Obesity_Type_II", "NObeyesdad"] = 6
df.loc[df["NObeyesdad"] == "Obesity_Type_III", "NObeyesdad"] = 7
df.head(74)
```

```python
# check if all categorical data are numerical now - returns 0 means yes
np.argmin(df.applymap(np.isreal).all(1))
```

```python
# normalization
label = df.NObeyesdad
df = df.drop(columns=["NObeyesdad"])
df = (df-df.min())/(df.max()-df.min())
df["NObeyesdad"] = label
print(df.head())
df.to_csv('Processed_data.csv',index = False)
```
